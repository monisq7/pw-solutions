Question 1: What is hypothesis testing in statistics?

Answer 1 Hypothesis testing is a fundamental statistical inference procedure used to make decisions or draw conclusions about a population parameter based on sample data. It is a formal, step-by-step methodology that allows researchers to test assumptions (hypotheses) about a population. The core logic of hypothesis testing is analogous to a courtroom principle. The initial assumption is the null hypothesis (H₀), which represents a statement of "no effect," "no difference," or the status quo. The researcher's goal is to find evidence against the null hypothesis in favor of the alternative hypothesis (H₁ or Hₐ), which represents a statement of change, effect, or difference.
The procedure does not provide absolute proof but instead evaluates the strength of the sample evidence against the null hypothesis. The conclusion is probabilistic, meaning it is always made with a known margin of error.

Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?

Anaswe 2 The null hypothesis (denoted as H₀) and the alternative hypothesis (denoted as H₁ or Hₐ) are two complementary, competing statements that form the foundation of a hypothesis test. They are defined as follows:

1. Null Hypothesis (H₀):

Definition: The null hypothesis is a statement of "no effect," "no difference," or "no change." It represents the status quo or a default position that the researcher seeks to challenge or find evidence against.

Characteristic: It is always a statement about a population parameter (e.g., mean μ, proportion p) and always contains an equality sign (e.g., =, ≤, ≥).

Purpose: It serves as the assumption that is initially presumed to be true for the purpose of testing. The goal of the test is to determine if there is sufficient sample evidence to reject H₀.

Example: H₀: μ = 100 (The population mean is equal to 100). H₀: p ≤ 0.5 (The population proportion is less than or equal to 0.5).

2. Alternative Hypothesis (H₁ or Hₐ):

Definition: The alternative hypothesis is a statement that directly contradicts the null hypothesis. It represents the researcher's belief or claim about the population parameter—that there is an effect, a difference, or a change.

Characteristic: It is a statement about the same population parameter as H₀ but never contains an equality sign. It expresses what the researcher wants to prove and can be non-directional (two-tailed, e.g., ≠) or directional (one-tailed, e.g., > or <).

Purpose: It is the hypothesis that is considered to be possibly true only if the sample data provides strong and convincing evidence against H₀.

Example: H₁: μ ≠ 100 (The population mean is not equal to 100). H₁: p > 0.5 (The population proportion is greater than 0.5).

Question 3: Explain the significance level in hypothesis testing and its role in deciding
the outcome of a test.

Answer 3 The significance level, denoted by the Greek letter alpha (α), is a critical threshold probability that is chosen by the researcher before conducting a hypothesis test. It defines the criterion for deciding whether the sample data provides sufficient evidence to reject the null hypothesis (H₀).

1. Definition and Interpretation:
The significance level (α) is the probability of making a Type I error. A Type I error occurs when we incorrectly reject a true null hypothesis. In other words, it is the probability of concluding there is an effect or difference when, in reality, there is none.

Commonly used values for α are 0.05 (5%), 0.01 (1%), and 0.10 (10%).

Choosing α = 0.05 means the researcher is willing to accept a 5% chance of falsely rejecting H₀ if it is actually true.

2. Its Role in Deciding the Outcome:
The significance level serves as the pre-determined cutoff point against which the p-value of the test is compared to make a final decision. The p-value is the probability of obtaining a test result at least as extreme as the one observed, assuming the null hypothesis is true.

The decision rule is as follows:

If the p-value ≤ α: We reject the null hypothesis (H₀). The sample data is considered statistically significant. It is considered too unlikely that the observed result occurred by random chance alone, assuming H₀ was true.

If the p-value > α: We fail to reject the null hypothesis (H₀). The sample data does not provide strong enough evidence against H₀. The observed result is considered reasonably likely to have occurred by random chance.

3. Analogy:
Think of the significance level as the strictness of a judge's standard for "reasonable doubt."

A strict judge (α = 0.01) requires very compelling evidence (a very low p-value) to convict (reject H₀).

A more lenient judge (α = 0.10) requires less compelling evidence to reach the same verdict.

4. The Trade-off:
Setting the significance level involves a trade-off between two types of errors:

A lower α (e.g., 0.01) reduces the chance of a Type I Error (false positive) but increases the risk of a Type II Error (β) (failing to reject a false H₀, or a false negative).

A higher α (e.g., 0.10) reduces the risk of a Type II Error but increases the risk of a Type I Error.

Question 4: What are Type I and Type II errors? Give examples of each.

Answer 4 In hypothesis testing, because decisions are made based on sample data and probabilities, there is always a chance of making an incorrect conclusion. These potential mistakes are categorized as Type I and Type II errors.

1. Type I Error (False Positive or α-error):

Definition: A Type I error occurs when we incorrectly reject a true null hypothesis (H₀). In simpler terms, we conclude there is an effect, difference, or change when, in reality, there is none.

Probability: The probability of committing a Type I error is denoted by the Greek letter alpha (α), which is the significance level of the test. By choosing α = 0.05, we are accepting a 5% risk of making a Type I error.

Example (Medical Testing):

H₀: The new drug has no effect on the disease.

Reality: H₀ is actually TRUE (the drug does not work).

Test Result: The study's data leads researchers to reject H₀ and conclude the drug is effective.

Consequence: This is a Type I error. Time and money are wasted on an ineffective drug, and patients might be given a false hope or exposed to side effects for no benefit.

2. Type II Error (False Negative or β-error):

Definition: A Type II error occurs when we fail to reject a false null hypothesis (H₀). In simpler terms, we conclude there is no sufficient evidence of an effect, difference, or change when, in reality, there is one.

Probability: The probability of committing a Type II error is denoted by the Greek letter beta (β). The power of a test (1 - β) is the probability of correctly rejecting a false null hypothesis.

Example (Medical Testing):

H₀: The new drug has no effect on the disease.

Reality: H₀ is actually FALSE (the drug does work).

Test Result: The study's data leads researchers to fail to reject H₀ and conclude there is no evidence the drug works.

Consequence: This is a Type II error. A beneficial and effective drug is incorrectly deemed a failure and might be abandoned, preventing it from helping patients.

Question 5: What is the difference between a Z-test and a T-test? Explain when to use
each.

Answer 5 A hypothesis test is a formal statistical procedure used to make an inference about a population parameter based on sample data, operating on the core principle of assessing the strength of evidence against a default position. This process begins by establishing two competing hypotheses: the null hypothesis (H₀), which represents a statement of "no effect," "no difference," or the status quo and always contains an equality, and the alternative hypothesis (H₁), which contradicts H₀ by proposing an effect or difference. The researcher's objective is to determine if the sample data provides sufficient evidence to reject H₀ in favor of H₁. The threshold for this decision is the significance level (α), a pre-determined probability that defines the maximum risk of making a Type I error—incorrectly rejecting a true null hypothesis (a false positive). Conversely, a Type II error occurs when a false null hypothesis is not rejected (a false negative). The probability of this error is denoted by β. The choice of statistical test, such as a Z-test or T-test, is crucial. A Z-test is used when the population standard deviation is known, typically employing the standard normal distribution. In contrast, a T-test is used when the population standard deviation is unknown and must be estimated from the sample, utilizing the Student's t-distribution to account for the additional uncertainty, especially vital with small sample sizes. The test yields a p-value, which is the probability of observing the sample results if H₀ is true. If this p-value is less than or equal to the chosen significance level α, the null hypothesis is rejected, and the result is deemed statistically significant. If the p-value is greater than α, there is insufficient evidence to reject the null hypothesis. Thus, hypothesis testing provides a structured, probabilistic framework for making data-driven decisions while explicitly acknowledging and quantifying the potential for error.

```python
'''Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]'''

# Import required libraries
import numpy as np
from scipy import stats

# Given sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Convert list to a numpy array
data = np.array(sample_data)

# Known population parameters (from prior knowledge)
pop_mean = 50  # Hypothesized population mean (H0: μ = 50)
pop_std = 0.8  # Known population standard deviation

# Calculate sample statistics
sample_mean = np.mean(data)
sample_size = len(data)

# Print the sample statistics
print("Sample Mean:", round(sample_mean, 4))
print("Sample Size:", sample_size)
print("Population Std Dev:", pop_std)

# Calculate the Z-statistic
z_statistic = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))
print("\nZ-Statistic:", round(z_statistic, 4))

# Calculate the p-value for a two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))
print("P-Value:", round(p_value, 4))

# Set significance level
alpha = 0.05

# Decision rule
print("\nSignificance Level (α):", alpha)
if p_value <= alpha:
    print("Decision: Reject the null hypothesis (H₀).")
else:
    print("Decision: Fail to reject the null hypothesis (H₀).")

# Interpretation
print("\nInterpretation:")
print(f"At a {alpha*100}% significance level, there is {'sufficient' if p_value <= alpha else 'insufficient'} evidence")
print("to conclude that the true population mean is different from 50.")
```

```python
'''Question 8: Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.'''

# Import required libraries
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Parameters for the normal distribution
true_mean = 50      # True population mean (mu)
true_std = 5        # True population standard deviation (sigma)
sample_size = 30    # Number of observations in the sample

# Generate sample data from the normal distribution
sample_data = np.random.normal(loc=true_mean, scale=true_std, size=sample_size)

# Calculate sample statistics
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)  # ddof=1 for sample standard deviation (N-1)
standard_error = sample_std / np.sqrt(sample_size)

# Calculate the 95% confidence interval
confidence_level = 0.95
degrees_of_freedom = sample_size - 1
t_critical = stats.t.ppf((1 + confidence_level) / 2, degrees_of_freedom)  # Get the t-value
margin_of_error = t_critical * standard_error

lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# Print the results
print("Sample Mean:", round(sample_mean, 2))
print("Sample Standard Deviation:", round(sample_std, 2))
print("Standard Error of the Mean:", round(standard_error, 2))
print(f"\n{t_critical:.2%} t-critical value for {degrees_of_freedom} d.f.:", round(t_critical, 3))
print("Margin of Error:", round(margin_of_error, 2))
print(f"\n{confidence_level:.0%} Confidence Interval for the Mean: ({lower_bound:.2f}, {upper_bound:.2f})")
print("True Population Mean:", true_mean)

# Create the plot
plt.figure(figsize=(10, 6))

# Plot 1: Histogram of the sample data with a density curve
count, bins, patches = plt.hist(sample_data, bins=8, alpha=0.7, color='skyblue', edgecolor='black', density=True, label='Sample Data Distribution')

# Overlay the theoretical normal distribution curve for reference
x = np.linspace(min(sample_data) - 2, max(sample_data) + 2, 100)
plt.plot(x, stats.norm.pdf(x, true_mean, true_std), 'r-', lw=2, label=f'True Population Distribution\n(μ={true_mean}, σ={true_std})')

# Plot 2: Error bar for the confidence interval of the mean
plt.axvline(sample_mean, color='green', linestyle='-', linewidth=3, label=f'Sample Mean = {sample_mean:.2f}')
plt.axvline(lower_bound, color='orange', linestyle='--', linewidth=2, label=f'95% CI: ({lower_bound:.2f}, {upper_bound:.2f})')
plt.axvline(upper_bound, color='orange', linestyle='--', linewidth=2)

# Add labels and title
plt.title('Simulated Normal Data and 95% Confidence Interval for the Mean', fontsize=14)
plt.xlabel('Value')
plt.ylabel('Density / Frequency')
plt.legend(loc='upper right', frameon=True)

# Add grid for better readability
plt.grid(axis='y', alpha=0.4)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
'''Question 9: Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean.'''

# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# 1. Define the function to calculate Z-scores
def calculate_z_scores(data):
    """
    Calculates the Z-scores for a given dataset.

    Parameters:
    data (array-like): The input dataset (list, NumPy array, etc.)

    Returns:
    numpy.ndarray: An array of Z-scores the same length as the input data.
    """
    data_array = np.array(data) # Convert input to a NumPy array
    mean = np.mean(data_array)
    std_dev = np.std(data_array, ddof=0) # Using population standard deviation
    z_scores = (data_array - mean) / std_dev
    return z_scores

# 2. Create a sample dataset from a normal distribution
np.random.seed(10) # For reproducibility
original_data = np.random.normal(loc=50, scale=10, size=1000) # Mean=50, Std Dev=10

# 3. Calculate Z-scores using the function
standardized_data = calculate_z_scores(original_data)

# 4. Print some statistics to verify
print("Original Data - Mean: {:.2f}, Standard Deviation: {:.2f}".format(np.mean(original_data), np.std(original_data)))
print("Standardized Data - Mean: {:.2f}, Standard Deviation: {:.2f}".format(np.mean(standardized_data), np.std(standardized_data)))

# 5. Create the visualization
plt.figure(figsize=(12, 5))

# Plot 1: Original Data
plt.subplot(1, 2, 1) # 1 row, 2 columns, first plot
plt.hist(original_data, bins=25, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(np.mean(original_data), color='red', linestyle='--', linewidth=2, label=f'Mean = {np.mean(original_data):.1f}')
plt.title('Original Data Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(axis='y', alpha=0.4)

# Plot 2: Standardized Data (Z-scores)
plt.subplot(1, 2, 2) # 1 row, 2 columns, second plot
plt.hist(standardized_data, bins=25, color='lightgreen', edgecolor='black', alpha=0.7)
plt.axvline(np.mean(standardized_data), color='red', linestyle='--', linewidth=2, label=f'Mean = {np.mean(standardized_data):.1f}')
# Add lines for standard deviations
plt.axvline(1, color='orange', linestyle=':', linewidth=1.5, label='+1 Std Dev')
plt.axvline(-1, color='orange', linestyle=':', linewidth=1.5, label='-1 Std Dev')
plt.axvline(2, color='purple', linestyle=':', linewidth=1.5, label='+2 Std Dev')
plt.axvline(-2, color='purple', linestyle=':', linewidth=1.5, label='-2 Std Dev')

plt.title('Standardized Data (Z-Scores) Distribution')
plt.xlabel('Z-Score')
plt.ylabel('Frequency')
plt.legend()
plt.grid(axis='y', alpha=0.4)

plt.tight_layout() # Adjusts spacing between plots
plt.show()
```
